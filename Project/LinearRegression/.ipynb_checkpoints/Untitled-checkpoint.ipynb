```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from nose.tools import *
import seaborn as sns

from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LinearRegression

from sklearn.model_selection import train_test_split
```

```python
sns.set(style = 'darkgrid')
sns.set_palette('deep')
```

```python
book_data = pd.read_csv("../data/books.csv", error_bad_lines=False)
book_data.head()
```

```python
sns.pairplot(book_data)
```

```python
sns.kdeplot(book_data.average_rating, shade= True)
plt.title('Rating Distribution\n')
plt.xlabel('Rating')
plt.ylabel('Frequency')
```

```python
book_data = book_data.drop()
```

```python
plt.figure(figsize=(15,15))
sns.boxplot(x = book_data['# num_pages'])
```

```python
plt.figure(figsize=(15,15))
sns.boxplot(x = book_data.text_reviews_count)
```

```python
book_data = book_data.drop(book_data.index[book_data['# num_pages'] >= 1500])
```

```python
book_data = book_data.drop(book_data.index[book_data['ratings_count'] >= 1000000])
```

```python
book_data = book_data.drop(book_data.index[book_data['text_reviews_count'] >= 20000])
```

```python
book_data.head()
```

```python
book_data.title = LabelEncoder().fit_transform(book_data.authors)
```

```python
book_data.authors = LabelEncoder().fit_transform(book_data.authors)
```

```python
book_data.head()
```

```python
book_data_features = book_data.drop(["average_rating", "isbn", "isbn13", "language_code"], axis=1)
```

```python
book_data_features.head()
```

```python
book_data_labels = book_data.average_rating
book_data_labels.head()
```

```python
book_data_features_train, book_data_features_test, book_data_labels_train, book_data_labels_test = train_test_split(book_data_features, book_data_labels, test_size=0.25)
```

```python
regression_model = LinearRegression()
regression_model.fit(book_data_features_train, book_data_labels_train)
```

```python
regression_model.score(book_data_features_test, book_data_labels_test)
```

```python
predictions = regression_model.predict(book_data_features_test)
pred = pd.DataFrame({'Actual': book_data_labels_test.tolist(), 'Predicted': predictions.tolist()}).head(25)
pred.head(10)
```
